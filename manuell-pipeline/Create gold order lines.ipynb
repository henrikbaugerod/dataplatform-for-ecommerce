```python
%sql
CREATE OR REPLACE TABLE gold_order_lines AS
SELECT o.order_id, o.customer_id, o.order_date, o.status, o.item_id, o.quantity, o.quantity * i.price revenue
FROM (
  SELECT order_id, customer_id, order_date, status, item.item_id, item.quantity
  FROM gold_orders
  LATERAL VIEW explode(items) AS item
) o
LEFT JOIN silver_items i
  ON o.item_id = i.item_id;

SELECT *
FROM gold_order_lines;
```
