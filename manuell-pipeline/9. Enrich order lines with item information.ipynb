##### Task 9
Enrich order lines with item information.

What I want in the end:
- An enriched table with product details joined in
- Columns including:
  - `order_id`
  - `customer_id`
  - `item_id`
  - `product_name`
  - `category`
  - `price`
  - `quantity`
  - line revenue
- One row per valid order line

```python
%sql
USE SCHEMA e_commerce;
```

```python
%sql
-- An enriched table with product details joined in

CREATE OR REPLACE TEMP VIEW silver_enriched_orders AS
SELECT o.order_id, o.customer_id, o.item_id, i.product_name, i.category, i.price, o.quantity
FROM (
  SELECT order_id, customer_id, item.item_id item_id, item.quantity quantity
  FROM silver_orders
  LATERAL VIEW explode(items) AS item
) o
LEFT JOIN silver_items i
  ON o.item_id = i.item_id;

SELECT *
FROM silver_enriched_orders;
```
