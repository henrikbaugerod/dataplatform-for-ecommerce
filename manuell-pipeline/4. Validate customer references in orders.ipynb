##### Task 4
Validate customer references in orders.

What I want in the end:
- A table of orders whose customer_id does not exist in customers
- Count of invalid customer references
- List of affected order_id values

```python
%sql
USE SCHEMA e_commerce;
```

```python
%sql
-- A table of orders whose customer_id does not exist in customers

SELECT o.*
FROM bronze_orders o
LEFT JOIN bronze_customers c
  ON o.customer_id = c.customer_id
WHERE c.customer_id IS NULL;
```

```python
%sql
-- Count of invalid customer references

SELECT COUNT(*) AS invalid_customer_references
FROM bronze_orders o
LEFT JOIN bronze_customers c
  ON o.customer_id = c.customer_id
WHERE c.customer_id IS NULL;
```

```python
%sql
-- List of affected order_id values

SELECT o.order_id AS affected_order_id
FROM bronze_orders o
LEFT JOIN bronze_customers c
  ON o.customer_id = c.customer_id
WHERE c.customer_id IS NULL;
```
