Here are Databricks tasks tailored to your three one-line JSON files and the intentional data-quality issues.

##### Task 1
Ingest `customers.json`, `items.json`, and `orders.json` into three raw tables.

What I want in the end:
- `bronze_customers`
- `bronze_items`
- `bronze_orders`
- All records loaded from the raw JSON files
- Nested `items` array preserved in `bronze_orders`

##### Task 2
Profile the raw data.

What I want in the end:
- A summary showing row counts for each table
- A summary of null values by column
- A summary of distinct key counts for:
  - `customer_id`
  - `item_id`
  - `order_id`

##### Task 3
Find duplicate business keys.

What I want in the end:
- A result set of duplicate `customer_id` values in customers
- A result set of duplicate `item_id` values in items
- A result set of duplicate `order_id` values in orders
- A count of duplicate rows per key

##### Task 4
Validate customer references in orders.

What I want in the end:
- A table of orders whose `customer_id` does not exist in customers
- Count of invalid customer references
- List of affected `order_id` values

##### Task 5
Validate item references inside orders.

What I want in the end:
- A flattened view of order lines from the nested `items` array
- A table of order lines whose `item_id` does not exist in items
- Count of invalid item references
- List of affected `order_id` and `item_id` pairs

##### Task 6
Validate data types and formats.

What I want in the end:
- A table of orders with invalid `order_date`
- A table of items with invalid or missing `price`
- A table of order lines with invalid `quantity`
- Counts for each validation failure type

##### Task 7
Build cleaned silver tables.

What I want in the end:
- `silver_customers`
- `silver_items`
- `silver_orders`
- No duplicate primary keys
- Valid dates in orders
- Valid foreign key references
- Valid item prices
- Valid nested order items

##### Task 8
Create a flattened order-lines table.

What I want in the end:
- One row per ordered item
- Columns including:
  - `order_id`
  - `customer_id`
  - `order_date`
  - `status`
  - `item_id`
  - `quantity`
- Saved as a reusable table, for example `silver_order_lines`

##### Task 9
Enrich order lines with item information.

What I want in the end:
- An enriched table with product details joined in
- Columns including:
  - `order_id`
  - `customer_id`
  - `item_id`
  - `product_name`
  - `category`
  - `price`
  - `quantity`
  - line revenue
- One row per valid order line

##### Task 10
Recalculate order totals.

What I want in the end:
- A table with one row per order
- Columns including:
  - `order_id`
  - recalculated total amount
  - total number of distinct items
  - total quantity ordered

##### Task 11
Build a customer summary table.

What I want in the end:
- One row per customer
- Columns including:
  - `customer_id`
  - `name`
  - `country`
  - number of valid orders
  - total items purchased
  - total revenue
  - average order value

##### Task 12
Build an item performance table.

What I want in the end:
- One row per item
- Columns including:
  - `item_id`
  - `product_name`
  - total quantity sold
  - total revenue
  - number of distinct orders containing the item

##### Task 13
Build a gold orders table.

What I want in the end:
- One row per order
- Fully enriched with customer data
- Includes aggregated order metrics
- Only valid, trusted records
- Saved as something like `gold_orders`

##### Task 14
Build a gold customer table.

What I want in the end:
- One row per customer
- Business-ready metrics for reporting
- Includes customers with zero valid orders if present
- Saved as something like `gold_customers`

##### Task 15
Create a data quality report.

What I want in the end:
- A single report or table showing:
- duplicate customers
- duplicate items
- duplicate orders
- missing customers in orders
- missing items in order arrays
- invalid dates
- null prices
- Each issue type should have a count and affected records

##### Task 16
Answer business questions from the cleaned data.

What I want in the end:
- Total revenue
- Revenue by country
- Revenue by order status
- Top 3 customers by revenue
- Top 3 items by quantity sold

##### Task 17
Separate bad records from good records.

What I want in the end:
- “Good” tables used for analytics
- “Quarantine” tables for rejected records
- A clear reason column explaining why each rejected record failed validation

###### Task 18
Document the transformation layers.

What I want in the end:

  - what belongs in bronze

##### Task 19
Make the pipeline idempotent.

What I want in the end:

- A short explanation of how duplicate keys are handled

##### Task 20
Prepare the dataset for dashboarding.

What I want in the end:
- At least two final reporting tables optimized for BI use
- Clear grain for each table
- Clean column names suitable for dashboards
