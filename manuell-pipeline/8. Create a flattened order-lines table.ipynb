##### Task 8
Create a flattened order-lines table.

What I want in the end:
- One row per ordered item
- Columns including:
  - `order_id`
  - `customer_id`
  - `order_date`
  - `status`
  - `item_id`
  - `quantity`
- Saved as a reusable table, for example `silver_order_lines`

```python
%sql
USE SCHEMA e_commerce;
```

```python
%sql
-- Create a flattened order-lines table

CREATE OR REPLACE TABLE silver_order_lines AS
SELECT order_id, customer_id, order_date, status, item.item_id item_id, item.quantity quantity
FROM silver_orders
LATERAL VIEW explode(items) AS item;

SELECT *
FROM silver_order_lines;
```
