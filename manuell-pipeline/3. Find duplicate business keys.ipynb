##### Task 3
Find duplicate business keys.

What I want in the end:
- A result set of duplicate customer_id values in customers
- A result set of duplicate item_id values in items
- A result set of duplicate order_id values in orders
- A count of duplicate rows per key

```python
%sql
-- A result set of duplicate customer_id values in customers
SELECT customer_id, COUNT(customer_id) total
FROM bronze_customers
GROUP BY customer_id
HAVING total > 1;
```

```python
%sql
-- A result set of duplicate item_id values in items
SELECT item_id, COUNT(item_id) total
FROM bronze_items
GROUP BY item_id
HAVING total > 1;
```

```python
%sql
-- A result set of duplicate order_id values in orders
SELECT order_id, COUNT(order_id) total
FROM bronze_orders
GROUP BY order_id
HAVING total > 1;
```

```python
%sql
-- A count of duplicate rows per key
SELECT *
FROM(
  SELECT 'customer_id' AS row_id, customer_id AS id, COUNT(customer_id) AS total FROM bronze_customers GROUP BY customer_id
  UNION ALL
  SELECT 'item_id', item_id, COUNT(item_id) FROM bronze_items GROUP BY item_id
  UNION ALL
  SELECT 'order_id', order_id, COUNT(order_id) FROM bronze_orders GROUP BY order_id
) t
WHERE total > 1
```
