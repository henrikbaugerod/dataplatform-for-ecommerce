##### Task 10
Recalculate order totals.

What I want in the end:
- A table with one row per order
- Columns including:
  - `order_id`
  - recalculated total amount
  - total number of distinct items
  - total quantity ordered

```python
%sql
USE SCHEMA e_commerce;
```

```python
%sql
-- A table with one row per order

SELECT o.order_id, SUM(i.price) total_price, COUNT(DISTINCT(o.item_id)) distinct_items, SUM(o.quantity) total_quantity
FROM (
  SELECT order_id, item.item_id, item.quantity
  FROM silver_orders
  LATERAL VIEW explode(items) AS item
) o
LEFT JOIN silver_items i
  ON o.item_id = i.item_id
GROUP BY order_id;
```
