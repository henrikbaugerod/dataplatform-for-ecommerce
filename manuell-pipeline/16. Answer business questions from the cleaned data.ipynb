##### Task 16
Answer business questions from the cleaned data.

What I want in the end:
- Total revenue
- Revenue by country
- Revenue by order status
- Top 3 customers by revenue
- Top 3 items by quantity sold

```python
%sql
USE SCHEMA e_commerce;
```

```python
%sql
CREATE OR REPLACE TABLE report_total_revenue AS
SELECT SUM(revenue) total_revenue
FROM gold_order_lines;

CREATE OR REPLACE TABLE report_revenu_by_country AS
SELECT c.country, SUM(o.revenue) revenue
FROM gold_order_lines o
LEFT JOIN gold_customers c
  ON o.customer_id = c.customer_id
GROUP BY c.country
ORDER BY revenue DESC;

CREATE OR REPLACE TABLE report_revenue_by_status AS
SELECT status, SUM(revenue) revenue
FROM gold_order_lines
GROUP BY status
ORDER BY revenue DESC;

CREATE OR REPLACE TABLE report_top_customers AS
SELECT o.customer_id, c.name, SUM(o.revenue) revenue
FROM gold_order_lines o
LEFT JOIN gold_customers c
  ON o.customer_id = c.customer_id
GROUP BY o.customer_id, c.name
ORDER BY revenue DESC, o.customer_id ASC
LIMIT 3;

CREATE OR REPLACE TABLE report_top_items AS
SELECT o.item_id, i.product_name, SUM(quantity) quantity
FROM gold_order_lines o
LEFT JOIN silver_items i
  ON o.item_id = i.item_id
GROUP BY o.item_id, i.product_name
ORDER BY quantity DESC, o.item_iD ASC
LIMIT 3;

SELECT *
FROM report_total_orders;
```

```python
%sql
CREATE OR REPLACE TABLE report_revenue_over_time AS
SELECT order_id, order_date, YEAR(order_date) year, MONTH(order_date) month, DAY(order_date) day, total_value
FROM gold_orders;
```
