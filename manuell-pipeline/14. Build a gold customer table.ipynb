Build a gold customer table.

What I want in the end:
- One row per customer
- Business-ready metrics for reporting
- Includes customers with zero valid orders if present
- Saved as something like `gold_customers`

```python
%sql
USE SCHEMA e_commerce;
```

```python
%sql
-- Build a gold customer table with one row per customer

CREATE OR REPLACE TABLE gold_customers AS
SELECT 
  c.customer_id, c.name, c.email, c.country, c.created_at,
  COUNT(o.order_id) total_orders,
  COALESCE(SUM(o.total_value), 0) total_revenue,
  COALESCE(AVG(o.total_value), 0) avg_order_value,
  MAX(o.order_date) latest_order
FROM silver_customers c
LEFT JOIN gold_orders o
  ON c.customer_id = o.customer_id
GROUP BY c.customer_id, c.name, c.email, c.country, c.created_at
ORDER BY c.customer_id;

SELECT *
FROM gold_customers;
```
